```python
class Prvek:
    def __init__(self, h, d=None):
        self.hodnota = h
        self.dalsi = d


class LSS:
    def __init__(self):
        self.zac = None

    def PridejHodnotuNaZacatek(self, h):
        self.zac = Prvek(h, self.zac)

    def OtocSeznam(self):
        """Otočí seznam do opačného pořadí
        POZOR: nechceme vytvářet žádné nové prvky!!!
        Metodu PridejHodnotuNaZacatek() nemůžeme volat, protože
        tato metoda vytváři nové instance třídy Prvek!
        """
        #pomocne promenne, kterymi menim odkazy u prvku v lss
        zadni=None #ukazuje dozadu
        predni=None #ukazuje dopredu a zachovava odkaz
        stavajici= self.zac
        while stavajici!=None:#vyradi mi i pripad, kde mam prazdny seznam/ nemusim delat podminku if self.zac=None:
          predni=stavajici.dalsi
          stavajici.dalsi=zadni
          zadni=stavajici
          stavajici=predni
        self.zac=zadni #da mi hlavu na posledni prvek

        #pass

    def VypustPrvek(self, p):
        """Vynechá ze seznamu prvek p (p je odkaz na prvek)
        p může byt None, nebo p nemusí být prvek v seznamu.
        """
        #if self.zac == None:
          #  return
        if (self.zac)==None:
          return(None)

        stavajici = self.zac
        previous=None #pomocna promenna do ktere si ulozim predchozi prvek

        while stavajici!=None :# proiteruje mi spojovy seznam dokud nenajdu odkaz na p
          if stavajici.hodnota==p:
            previous.dalsi=stavajici.dalsi.dalsi#nahradim odkaz na p odkazem na prvek za nim
            return
          previous=stavajici
          stavajici=stavajici.dalsi
        return


       # funguje pro =, nefunguje pro <

    def VypustPrvkySHodnotou(self, h):
        """Vynechá ze seznamu všechny prvky s hodnotou h
        jsou-li tam nějaké.
        """
        if (self.zac)==None:
          return(None)
        if self.zac.dalsi==None and h==self.zac.hodnota:
          return(None)
        stavajici = self.zac

        while stavajici.dalsi!=None:# proiteruje mi spojovy seznam
          if stavajici.dalsi==h:#pokazde co najdu odkaz na h tak
            stavajici.dalsi=stavajici.dalsi.dalsi #nahradim odkaz na h odkazem na prvek za nim
          stavajici=stavajici.dalsi
        return ()
        #nefunguje

    def VypustPrvkySHodnotouVetsiNez(self, h):
        """Vynechá ze seznamu všechny prvky s hodnotou ostře větší než h
        jsou-li tam nějaké.
        """
        if (self.zac)==None:
          return(None)

        if self.zac.dalsi==None and self.zac.hodnota > h:
          return(None)
        stavajici = self.zac

        while stavajici.dalsi!=None :# proiteruje mi spojovy seznam
          if stavajici.dalsi.hodnota > h:#pokazde co najdu odkaz na neco ostre vesti nez h tak
              stavajici.dalsi=stavajici.dalsi.dalsi
              if stavajici.dalsi==None:
                return() #nahradim odkaz na to neco odkazem na prvek za nim
          stavajici=stavajici.dalsi
        return ()
       # udelalo jen pro jeden nebo dva z peti, jeden ze tri a pro dva nene type

    def NajdiPrvekSHodnotou(self, h):
        """Hledá v seznamu první (od začátku) prvek s hodnotou h.
        Když tam je alespoň jeden prvek s hodnotou h, tak vrátí ten,
        který je nejblíže začátku seznamu.
        Když prvek s hodnotou h v seznamu není, tak funkce vrátí None.
        """
        if self.zac == None:
            return(None)
        stavajici=self.zac

        while stavajici!=None:# proiteruje mi spojovy seznam dokud nenajdu odkaz na prvek s hodnotou h
          if stavajici.hodnota==h:
            return(stavajici) #nahradim odkaz na p odkazem na prvek za nim
          stavajici=stavajici.dalsi
        return(None)
        #funguje

    def NajdiPrvekSHodnotouVetsiNez(self, h):
        """Hledá v seznamu nejmenší prvek (od začátku) s hodnotou větší než h.
        Vrátí prvek s nejmenší hodnotou, která je větší než h.
        Když je tam takových více, tak
        ten, který je nejblíže začátku seznamu.
        Když seznam neobsahuje žádné prvky s hodnotou větší h,
        tak funkce vrátí None.
        """
        if self.zac == None:
            return (None)

        stavajici = self.zac
        nejmensi=None

        while stavajici!=None:
          if stavajici.hodnota> h: #je li stavajici prvek ostre vetsi nez h , zapamatuje si ho pomocna promenna nejmensi
            if nejmensi==None or (stavajici.hodnota < nejmensi.hodnota): #je li prvek ostre mensi nez dosavadni nejmensi vetsi prvek, nahradi se hodnota
              nejmensi=stavajici
          stavajici=stavajici.dalsi
        return(nejmensi)
        #funguje

    def Vypis(self):
        """vypise hodnoty ze sezanmu na jeden radek"""
        pom = self.zac
        while pom:
            print(pom.hodnota, end=' ')
            pom = pom.dalsi
        print('')       # konec radku

s = LSS()
pocet_prikazu = int(input())
for _ in range(pocet_prikazu):
    radek = input().split()
    prikaz = radek[0]
    parametry = radek[1:]
    if prikaz == 'OtocSeznam':
        s.OtocSeznam()
    elif prikaz == 'PridejHodnotuNaZacatek':
            s.PridejHodnotuNaZacatek(int(parametry[0]))
    elif prikaz == 'VypustPrvek':
        if parametry[0] == '=':
            p = s.NajdiPrvekSHodnotou(int(parametry[1]))
            s.VypustPrvek(p)
        elif parametry[0] == '>':
            p = s.NajdiPrvekSHodnotouVetsiNez(int(parametry[1]))
            s.VypustPrvek(p)
        else:
            print(f'Neznamy parametr v prikazu "{prikaz} {parametry[0]}"')
    elif prikaz == 'VypustPrvkySHodnotou':
        s.VypustPrvkySHodnotou(int(parametry[0]))
    elif prikaz == 'VypustPrvkySHodnotouVetsiNez':
        s.VypustPrvkySHodnotouVetsiNez(int(parametry[0]))
    else:
        print(f'Neznamy prikaz "{prikaz}"',)
    s.Vypis()

'''15
PridejHodnotuNaZacatek 0
VypustPrvkySHodnotou 10
VypustPrvek = 0
VypustPrvek > 0
PridejHodnotuNaZacatek 2
PridejHodnotuNaZacatek -5
PridejHodnotuNaZacatek 8
PridejHodnotuNaZacatek 2
OtocSeznam
VypustPrvkySHodnotouVetsiNez 5
VypustPrvkySHodnotou 2
PridejHodnotuNaZacatek 1
VypustPrvkySHodnotou -3
VypustPrvek > 0
VypustPrvek = 9'''
```
